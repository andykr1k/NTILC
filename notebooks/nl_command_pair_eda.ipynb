```python
import json

INPUT_FILE_PATH = '../data/man/nl_command_pairs.json'
OUTPUT_FILE_PATH = '../data/man/nl_command_pairs_cleaned.json'
```

```python
with open(INPUT_FILE_PATH, 'r') as file:
    data = json.load(file)
```

```python
data
```

```python
results = [obj for obj in data if obj.get("examples") is not None]
```

```python
len(results)
```

```python
print([obj for obj in results if obj.get("tool") == 'cat'])
```

```python
with open(OUTPUT_FILE_PATH, 'w', encoding='utf-8') as json_file:
    json.dump(results, json_file, indent=4)
```
